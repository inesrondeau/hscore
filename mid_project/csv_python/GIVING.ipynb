```python
import imblearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
giving_df=pd.read_csv('giving.txt', delimiter= "\s+",engine='python')
```

```python
giving_df
```

```python
import re
```

```python
clean=[]
for value in giving_df['Country']:
    pattern= '_'
    clean.append(re.sub(pattern, ' ', value))
     
```

```python
giving_df['Country name']=clean
```

```python
giving_df=giving_df.drop(['Country','Ranking_hs','Ranking.1','Ranking_v','Ranking'],axis=1)
```

```python
giving_df
```

```python
giving_df=giving_df[['Country name', 'Score(%)_hs', 'Score(%)_d', 'Score(%)_v']]
```

```python
giving_df
```

```python
def rename(x):
    if x=='Russian Federation':
        return 'Russia'
    elif x=='Republic of Moldova':
        return 'Moldova'
    elif x=='United Republic of Tanzania':
        return 'Tanzania'
    else:
        return x
```

```python
giving_df['Country name']=giving_df['Country name'].apply(rename)
```

```python
giving_df.to_csv('GIVING.csv')
```
