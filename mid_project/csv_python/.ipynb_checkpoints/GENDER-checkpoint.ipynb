```python
import imblearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
gender_df=pd.read_csv('gender1.csv')
```

```python
gender_df=gender_df[0:-5]
```

```python
gender_df['Series Name'].unique()
```

```python
grouper1 = gender_df.groupby(['Country Name'])
gender= pd.concat([pd.Series(v['2015 [YR2015]'].tolist(), name=k) for k, v in grouper1], axis=1).T
```

```python
gender.columns=gender_df['Series Name'].unique()
```

```python
gender
```

```python
def points (x):
    if x=='..':
        return np.NAN
    else:
        return x
    
```

```python
for column in gender:
    gender[column]=gender[column].apply(points)
```

```python
gender.isna().sum()/len(gender)
```

```python
data_hscore=pd.read_excel('DataPanelWHR2021C2.xls')
```

```python
data_hscore=data_hscore[(data_hscore['year']==2015)]
```

```python
countries=list(data_hscore['Country name'])
```

```python
gender=gender.reset_index()
```

```python
gender=gender.rename(columns={'index':'Country name'})
```

```python
list(gender['Country name'])
```

```python
def rename(x):
    if x=='Egypt, Arab Rep.':
        return 'Egypt'
    elif x=='Hong Kong SAR, China':
        return 'Hong Kong S.A.R. of China'
    elif x=='Iran, Islamic Rep.':
        return 'Iran'
    elif x=='Kyrgyz Republic':
        return 'Kyrgyzstan'
    elif x=='Russian Federation':
        return 'Russia'
    elif x=='Slovak Republic':
        return 'Slovakia'
    elif x=='Syrian Arab Republic':
        return 'Syria'
    elif x=='Venezuela, RB':
        return 'Venezuela'
    elif x=='Yemen, Rep.':
        return 'Yemen'
    else:
        return x
```

```python
gender['Country name']=gender['Country name'].apply(rename)
```

```python
def reduce (x):
    if x in countries:
        return x
    else:
        return 0
```

```python
gender['Country name']=gender['Country name'].apply(reduce)
```

```python
gender=gender[~(gender['Country name'] == 0)]
```

```python
gender
```

```python
gender=gender[gender['Adolescent fertility rate (births per 1,000 women ages 15-19)'].isna()==False]
gender=gender[gender['Law prohibits discrimination in employment based on gender (1=yes; 0=no)'].isna()==False]
```
