```python
import imblearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
suicide_df=pd.read_csv('crudeSuicideRates.csv')
```

```python
suicide_df=suicide_df[suicide_df['Period']==2015]
```

```python
suicide_df=suicide_df[suicide_df['Dim1']=='Both sexes']
```

```python
suicide_df=suicide_df.reset_index()
```

```python
suicide_df=suicide_df.drop(['index','Indicator','Dim1','Period'],axis=1)
```

```python
suicide_df
```

```python
list(suicide_df['Location'])
```

```python
def rename(x):
    if x=='Bolivia (Plurinational State of)':
        return 'Bolivia'
    elif x=='Czechia':
        return 'Czech Republic'
    elif x=='Iran (Islamic Republic of)':
        return 'Iran'
    elif x=='Russian Federation':
        return 'Russia'
    elif x=='Syrian Arab Republic':
        return 'Syria'
    elif x=='United Kingdom of Great Britain and Northern Ireland':
        return 'United Kingdom'
    elif x=='United States of America':
        return 'United State'
    elif x=='Venezuela (Bolivarian Republic of)':
        return 'Venezuela'
    elif x=='Viet Nam':
        return 'Vietnam'
    else:
        return x
```

```python
suicide_df['Location']=suicide_df['Location'].apply(rename)
```

```python
suicide_df.to_csv('SUICIDE.csv')
```
