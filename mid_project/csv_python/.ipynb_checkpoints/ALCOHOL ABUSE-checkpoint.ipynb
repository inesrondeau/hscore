```python
import imblearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
alcohol_df=pd.read_csv('alcoholSubstanceAbuse.csv')
```

```python
alcohol_df
```

```python
alcohol_df=alcohol_df[alcohol_df['Period']==2015]
alcohol_df=alcohol_df[alcohol_df['Dim1']=='Both sexes']
alcohol_df=alcohol_df.reset_index()
alcohol_df=alcohol_df.drop(['index','Indicator','Dim1','Period'],axis=1)
```

```python
list(alcohol_df['Location'].sort_values())
```

```python
def rename(x):
    if x=='Bolivia (Plurinational State of)':
        return 'Bolivia'
    elif x=='Czechia':
        return 'Czech Republic'
    elif x=='Iran (Islamic Republic of)':
        return 'Iran'
    elif x=='Russian Federation':
        return 'Russia'
    elif x=='Syrian Arab Republic':
        return 'Syria'
    elif x=='United Kingdom of Great Britain and Northern Ireland':
        return 'United Kingdom'
    elif x=='United States of America':
        return 'United State'
    elif x=='Venezuela (Bolivarian Republic of)':
        return 'Venezuela'
    elif x=='Viet Nam':
        return 'Vietnam'
    else:
        return x
```

```python
alcohol_df['Location']=alcohol_df['Location'].apply(rename)
```

```python
alcohol_df.to_csv('ALCOHOLS.csv')
```
